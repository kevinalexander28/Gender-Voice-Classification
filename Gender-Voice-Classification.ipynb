```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# read data
df = pd.read_csv("data/voice.csv", header=None)
df
```

# Data Pre-processing

```python
# change columns names
df.columns = ["meanfreq", "sd", "median", "Q25", "Q75", "IQR", "skew", "kurt", "sp.ent", "sfm", "mode", "centroid", "meanfun", "minfun", "maxfun", "meandom", "mindom", "maxdom", "dfrange", "modindx", "label"]
df
```

```python
# Clean the dataset
df.isnull().sum()
```

# Modeling 

### Split Data Train-Test 

```python
# import library
from sklearn.model_selection import train_test_split

# split data
X = df.iloc[:, 0:20]
# print(X)
y = df.iloc[:, 20]
# print(y)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=99)
```

# Decision Tree

```python
# import library
from sklearn.tree import DecisionTreeClassifier

# Fitting the decision tree with default hyperparameters, apart from
# max_depth which is 5 so that we can plot and read the tree.

# modeling
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

# K-Nearest Neighbor

```python
# import library
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# normalization
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

# determine k
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    y_pred_default = knn.predict(X_test)
    acctrain = knn.score(X_train, y_train)
    acctest = knn.score(X_test, y_test)
    print(i, ' ', acctrain, ' ', acctest)
    
knn = KNeighborsClassifier(n_neighbors = 3)
scores = cross_val_score(knn, X_train, y_train, cv = 10)
print("Cross Val Score:", scores)

print("Avg Cross Val Score:", np.average(scores))

# modeling
knn.fit(X_train, y_train)
```

# Evaluation

```python
# import library
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

# show accuracy
print("Training Accuracy =======================================")
print("DT Accuracy on training set: {:.3f}".format(dt.score(X_train, y_train))) 
print("KNN Accuracy on training set: {:.3f}".format(knn.score(X_train, y_train))) 
print("Testing Accuracy =======================================")
print("DT Accuracy on test set: {:.3f}".format(dt.score(X_test, y_test)))
print("KNN Accuracy on test set: {:.3f}".format(knn.score(X_test, y_test)))

# making predictions
y_pred_dt = dt.predict(X_test)
y_pred_knn = knn.predict(X_test)

# printing classifier report after prediction
print("Classification Report DT:\n", classification_report(y_test,y_pred_dt))
print("Classification Report KNN:\n", classification_report(y_test,y_pred_knn))

# printing confusion matrix and accuracy
print("Confusion Matrix DT:\n", confusion_matrix(y_test,y_pred_dt))
print("Confusion Matrix KNN:\n", confusion_matrix(y_test,y_pred_knn))
```
